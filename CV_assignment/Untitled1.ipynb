```python
import cv2
import os
import glob
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Define paths
input_folder = "data/test/images/d_r_47_.jpg"
output_folder = r"C:\Users\USER\Desktop\SUIM-master\data\test\filtered_images"
```

```python
# Create output folder if not exists
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Get list of image files (supports PNG and JPG)
image_files = glob.glob(os.path.join(input_folder, "*.png")) + glob.glob(os.path.join(input_folder, "*.jpg"))
```

```python
# Apply Gaussian Blur to each image
for image_path in image_files:
    image = cv2.imread(image_path)
    if image is None:
        continue  # Skip if the image couldn't be read
    
    blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
```

```python
# Save the filtered image
    filename = os.path.basename(image_path)
    output_path = os.path.join(output_folder, filename)
    cv2.imwrite(output_path, blurred_image)
```
